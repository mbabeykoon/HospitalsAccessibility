```python
#read county data into county_df DataFrame  from csv and disply first 5 rows.
county_df = pd.read_csv('../Data/uscounties.csv') 
county_df.head(5)
```

```python
import numpy as np
from geopy.distance import geodesic


# Precompute hospital lat-lng tuples for faster access
hospital_coords = [(lat, lng) for lat, lng in zip(hospital_df['LATITUDE'], hospital_df['LONGITUDE'])]

# Function to calculate the distance to the nearest hospital
def distance_to_nearest_hospital(county_center_lat, county_center_lng, hospital_coords):
    min_distance = float('inf')
    for hospital_lat, hospital_lng in hospital_coords:
        distance = geodesic((county_center_lat, county_center_lng), (hospital_lat, hospital_lng)).miles
        min_distance = min(min_distance, distance)
    return min_distance

# Apply the function on county_df
county_df['Max Distance to Hospital'] = county_df.apply(
    lambda x: distance_to_nearest_hospital(x['lat'], x['lng'], hospital_coords), axis=1)

# Result
print(county_df[['county', 'state_id', 'Max Distance to Hospital']])
county_df.to_csv('../outputs/max_distances_to_hospital.csv')
```

```python
# Load the shapefile ,  merge to max_distances_df and disply first 5 rows.


import geopandas as gpd

gdf = gpd.read_file('../shapefiles/cb_2018_us_county_500k.shp')


merged_gdf = gdf.merge(county_df, left_on='NAME', right_on='county')
merged_gdf.head(1)
```

```python
#plot hospital locations and max distance form center of each county to nearest hospital
import plotly.express as px
import plotly.graph_objects as go
import json



# Convert GeoDataFrame to JSON for the choropleth
json_data = merged_gdf.geometry.to_json()

# Convert GeoJSON to dictionary (which is compatible with plotly)
geojson = json.loads(json_data)

# Create the base choropleth map
fig = px.choropleth_mapbox(
    merged_gdf,
    geojson=geojson,
    locations=merged_gdf.index,  # Use the column with county names
    color='Max Distance to Hospital',  # Variable for coloring
    color_continuous_scale="Viridis",  # Color scale
    mapbox_style="carto-positron",
    zoom=3,
    center={"lat": 37.0902, "lon": -95.7129},
    opacity=0.5,
    hover_name='NAME',
)

# Add scatter plot for hospital locations
fig.add_trace(go.Scattermapbox(
    lat=hospital_df['LATITUDE'],
    lon=hospital_df['LONGITUDE'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=4,
        color='red',
        opacity=0.7
    ),
    
))

# Adjust color bar size
fig.update_layout(coloraxis_colorbar=dict(
    title='Max Distance to Hospital in each County',
    thicknessmode="pixels", thickness=10,
    lenmode="pixels", len=200
))

# Hide the axis ticks and labels
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

# Save the figure as an HTML file
fig.write_html("../outputs/HospitalAccessibilityFig1.html")
```
